```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.externals import joblib
from sklearn.metrics import accuracy_score
```

```python
from xgboost import XGBClassifier
```

```python
dataset = pd.read_csv('')
```

```python
dataset.head()
```

```python
df = dataset.drop(dataset.columns[0],axis=1)
```

```python
df.head()
```

```python
df= df.fillna(0)
```

```python
factor = pd.factorize(df['0.1'])
df['0.1'] = factor[0]
definitions = factor[1]
definitions
df.head()
```

```python
X = df.iloc[:,0:216].values
y = df.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 21)
```

```python
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
y_test
```

```python
accuracy_score(y_test,y_pred)
```

```python
reversefactor = dict(zip(range(10),definitions))
y_test = np.vectorize(reversefactor.get)(y_test)
y_pred = np.vectorize(reversefactor.get)(y_pred)
```

```python
y_test
```

```python
cm = pd.crosstab(y_test, y_pred, rownames=['Actual emotions'], colnames=['Predicted emotions'])
```
